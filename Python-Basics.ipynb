# Strings

```python
# String Literals

print("Hello")
print('Hello')

print("------------------------------------------------------------------")
# Assign String to a Variable
str1 = "Today"
print(str1)
print("------------------------------------------------------------------")

# Multiline Strings
str2 = """Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua."""
print(str2)
print("------------------------------------------------------------------")

#Strings are Arrays
str3 = "Welcome, Daniel!"
print(str3[1])
print(str3[2:5])
print("------------------------------------------------------------------")

# strip() method: The strip() method removes any whitespace from the beginning or the end:
str4 = " Hello, World! "
print(str4.strip())
print("------------------------------------------------------------------")

#  This example splits the string into substrings if it finds instances of the separator:
print(str4.split(",")) # returns ['Hello', ' World!']
print("------------------------------------------------------------------")

# Len() method: The len() method returns the length of a string:
print(len(str4))
print("------------------------------------------------------------------")

# Lower() Method: The lower() method returns the string in lower case:
print(str4.lower())
print("------------------------------------------------------------------")

# Upper() Method: The upper() method returns the string in upper case:
print(str4.upper())
print("------------------------------------------------------------------")

# Replace() method: The replace() method replaces a string with another string:
print(str4.replace("H", "J"))
print("------------------------------------------------------------------")

# String Format

#Example 1
# We can combine strings and numbers by using the format() method!
# Use the format() method to insert numbers into strings:
age1 = 36
txt = "My name is John, and I am {}"
print(txt.format(age1))
print("------------------------------------------------------------------")

#Example 2
quantity = 3
itemno = 567
price = 49.95
myorder = "I want {} pieces of item {} for {} dollars."
print(myorder.format(quantity, itemno, price))
print("------------------------------------------------------------------")

#You can use index numbers {0} to be sure the arguments are placed in the correct placeholders:
#Example 3
quantity1 = 3
itemno1 = 567
price1 = 49.95
myorder1 = "I want to pay {2} dollars for {0} pieces of item {1}."
print(myorder1.format(quantity, itemno, price))
print("------------------------------------------------------------------")

# Example 4
# we cannot combine strings and numbers like this:
age = 33
txt = "My name is John, I am " + age
print(txt)
print("------------------------------------------------------------------")
```

# Functions

```python
# Creating a Function

def my_function():
  print("This is a function")
print("------------------------------------------------------------------")

# Calling a Function
my_function()
print("------------------------------------------------------------------")

# Parameters
def function1(name):
  print(name + ":")

function1("Mehmet")
function1("Ali")
function1("John")
print("------------------------------------------------------------------")

# Default Parameter Value: If we call the function without parameter, it uses the default value:
def function2(country = "Turkey"):
  print("I am from " + country)

function2("Sweden")
function2("Saudi-Arabia")
function2()
function2("Germany")
print("------------------------------------------------------------------")


# Passing a List as a Parameter: You can send any data types of parameter to a function 
    # (string, number, list, dictionary etc.), and it will be treated as the same data type inside the function.
def function3(Months):
  for x in Months:
    print(x)

months_list = ["January", "February", "April"]

function3(months_list)
print("------------------------------------------------------------------")

# Return Values
def function4(x):
  return 5 * x

print(function4(2))
print(function4(6))
print(function4(10))
print("------------------------------------------------------------------")

# Recursion: Example-Python Program to Find Factorial of Number Using Recursion
def function5(val):
    if (val==1):
        return val
    else:
        return val*function5(val-1)

print(function5(5))
print("------------------------------------------------------------------")

# Flexible  Function:
age=[10,20]

def function6(length,weight,*args):
    print(args[0])
    output = (length+weight)*args[0]
    return output
print(function6(155,60,age[0],age[1]))
print("------------------------------------------------------------------")

# Lambda Function: The expression is executed and the result is returned:
x = lambda a : a + 25
print(x(5))

y = lambda a, b : a * b
print(y(5, 6))

z = lambda a, b, c : a + b + c
print(z(5, 6, 2))
print("------------------------------------------------------------------")

# Example 1: Calculating the circumference of a circle (Default Value)
def calculate(r,pi=3.14):
    output = 2*pi*r
    return output

print(calculate(2))
print("------------------------------------------------------------------")

# Example 2: Flexible Function
word1="Apple"
word2="Banana"
word3="Melon"

def function7(*args):
    for i in args:
        print(i)
        
function7(word1,word2,word3)
```

# Lists

```python
# Create a List:
thislist = ["Mercedes", "Volvo", "Audi"]
print(thislist)
print("------------------------------------------------------------------")

# Access Items
print(thislist[1])
print("------------------------------------------------------------------")

# Change Item Value
list1 = ["Mercedes", "Volvo", "Audi"]
list1[1] = "Opel"
print(list1)
print("------------------------------------------------------------------")

# Loop Through a List
thislist = ["Mercedes", "Volvo", "Audi"]
for x in thislist:
  print(x)
print("------------------------------------------------------------------")

# Check if Item Exists
mylist = ["apple", "banana", "cherry"]
if "apple" in mylist:
  print("Yes, 'apple' is in the fruits list")
print("------------------------------------------------------------------")

# List Length
print(len(mylist))
print("------------------------------------------------------------------")

# Add Items
mylist.append("melon")
print(mylist)
print("------------------------------------------------------------------")

# Remove Item
  # Method 1: The remove() method removes the specified item:
numberlist = [1, 2, 3]
numberlist.remove(1)
print(numberlist)

  # Method 2: The pop() method removes the specified index, (or the last item if index is not specified):
numberlist = [1, 2, 3]
numberlist.pop()
print(numberlist)

  # Method 3: The del keyword removes the specified index:
numberlist = [1, 2, 3]
del numberlist[0]
print(numberlist)

  # The del keyword can also delete the list completely:
numberlist = [1, 2, 3]
del numberlist

  # Method 4: The clear() method empties the list:
numberlist = [1, 2, 3]
numberlist.clear()
print(numberlist)

print("------------------------------------------------------------------")

# Copy a List
numberlist = [1, 2, 3]
numberlist2=numberlist.copy()
print(numberlist2)
```

# Dictionaries

```python
# A dictionary is a collection which is unordered, changeable and indexed. 
# In Python dictionaries are written with curly brackets, and they have keys and values.

# Create a dictionary:
thisdict ={
  "brand": "Opel",
  "model": "Corsa",
  "year": 2000
}
print(thisdict)
print("------------------------------------------------------------------")

# Accessing Items
x = thisdict["model"]
print(x)
# There is also a method called get() that will give you the same result:
print(thisdict.get("model"))

print("------------------------------------------------------------------")

# Change Values
thisdict["year"] = 2018
print(thisdict)
print("------------------------------------------------------------------")

# Loop Through a Dictionary
  # Print all key names in the dictionary, one by one:
for x in thisdict:
  print(x)
  # You can also use the keys() function to return keys of a dictionary:
for x in thisdict.keys():
  print(x)
  # Print all values in the dictionary, one by one:
for x in thisdict:
  print(thisdict[x])
  # You can also use the values() function to return values of a dictionary:
for x in thisdict.values():
  print(x)
  # Loop through both keys and values, by using the items() function:
for x, y in thisdict.items():
  print(x, y)
print("------------------------------------------------------------------")

# Dictionary Length
print(len(thisdict))
print("------------------------------------------------------------------")

# Adding Items
thisdict ={
  "brand": "Opel",
  "model": "Corsa",
  "year": 2000
}
thisdict["color"]="red"
print(thisdict)
print("------------------------------------------------------------------")

# Removing Items
  # The pop() method removes the item with the specified key name:
thisdict.pop("model")
print(thisdict)

  # The popitem() method removes the last inserted item (in versions before 3.7, a random item is removed instead):
thisdict ={
  "brand": "Opel",
  "model": "Corsa",
  "year": 2000
}
del thisdict["model"]
print(thisdict)

  # The del keyword can also delete the dictionary completely:
del thisdict
# print(thisdict) # this will cause an error because "thisdict" no longer exists.

  # The clear() keyword empties the dictionary:
thisdict ={
  "brand": "Opel",
  "model": "Corsa",
  "year": 2000
}
thisdict.clear()
print(thisdict)
```

Dictionary Methods

| Method | Description |

| copy() | Returns a copy of the dictionary |
| fromkeys() | Returns a dictionary with the specified keys and values |

# Python If ... Else

Python Conditions and If statements


| Method | Description |
| --- | --- |
| Equals | a == b |
| Not Equals | a != b |
| Less than | a < b |
| Less than or equal to | a <= b |
| Greater than | a > b |
| Greater than or equal to | a >= b |

```python
# If statement:
a = 330
b = 1500
if b > a:
  print("b is greater than a")
print("------------------------------------------------------------------")

# Elif: The elif keyword is pythons way of saying "if the previous conditions were not true, then try this condition".
a = 33
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
print("------------------------------------------------------------------")

# Else
a = 200
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is greater than b")
print("------------------------------------------------------------------")

# Short Hand If
if a > b: print("a is greater than b")
print("------------------------------------------------------------------")

# Short Hand If ... Else
print("A") if a > b else print("B")
print("A") if a < b else print("=") if a == b else print("B")
print("------------------------------------------------------------------")

# And
a = 150
b = 140
c = 300
if a > b and c > a:
  print("Both conditions are True")
print("------------------------------------------------------------------")

# Or
if a > b or a > c:
  print("At least one of the conditions is True")
print("------------------------------------------------------------------")
```

# For Loops

```python
days = ["Monday", "Tuesday", "Friday"]
for x in days:
  print(x)
print("------------------------------------------------------------------")

# The break Statement: With the break statement we can stop the loop before it has looped through all the items:
days = ["Monday", "Tuesday", "Friday"]
for x in days:
  print(x) 
  if x == "Monday":
    break
print("------------------------------------------------------------------")

# The continue Statement: With the continue statement we can stop the current iteration of the loop, and continue with the next:
days = ["Monday", "Tuesday", "Friday"]
for x in days:
  if x == "Monday":
    continue
  print(x)
print("------------------------------------------------------------------")

# The range() Function
for x in range(6): # Note that range(6) is not the values of 0 to 6, but the values 0 to 5.
  print(x) 
print("------------------------------------------------------------------")
for x in range(2, 6):
  print(x)
print("------------------------------------------------------------------")
for x in range(2, 30, 3):
  print(x)
print("------------------------------------------------------------------")

# Else in For Loop: The else keyword in a for loop specifies a block of code to be executed when the loop is finished:
for x in range(6):
  print(x)
else:
  print("Finally finished!")

# Nested Loops
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
  for y in fruits:
    print(x, y)
```

# Python While Loops

```python
# The while Loop
i = 1
while i < 10:
  print(i)
  i += 1
print("------------------------------------------------------------------")

# The break Statement: With the break statement we can stop the loop even if the while condition is true:
i = 1
while i < 15:
  print(i)
  if i == 5:
    break
  i += 1
print("------------------------------------------------------------------")

# The continue Statement: With the continue statement we can stop the current iteration, and continue with the next:
i = 0
while i < 15:
  i += 1 
  if i == 5:
    continue
  print(i)


```
